```python
import yfinance as yf
import json
import os
from datetime import datetime, timedelta

# Lista de 50 aÃ§Ãµes populares
most_popular_symbols = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "META", "TSLA", "NVDA", "NFLX", "INTC", "AMD",
    "BABA", "DIS", "PYPL", "ADBE", "CRM", "PEP", "KO", "NKE", "WMT", "PFE",
    "MRNA", "XOM", "CVX", "JNJ", "JPM", "BAC", "C", "WFC", "V", "MA",
    "T", "VZ", "TMO", "UNH", "LLY", "ABT", "BA", "GE", "GM", "F",
    "QCOM", "TXN", "ORCL", "IBM", "GS", "PLTR", "ROKU", "SQ", "UBER", "LYFT"
]

# Novo diretÃ³rio de saÃ­da
output_dir = r""
os.makedirs(output_dir, exist_ok=True)

# Intervalo de 10 anos atÃ© 2025-05-09
data_limite = datetime(2025, 5, 9)
dez_anos_atras = data_limite - timedelta(days=365*10)

# FunÃ§Ã£o para serializar DataFrames com datas
def df_to_records(df):
    df = df.copy()
    df.index = df.index.astype(str)
    return df.reset_index().to_dict(orient="records")

# FunÃ§Ã£o para salvar JSON
def save_stock_data(symbol, data):
    path = os.path.join(output_dir, f"{symbol}.json")
    with open(path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2)

# Coleta
falhas = []
for symbol in most_popular_symbols:
    print(f"ğŸ”„ Coletando dados para {symbol}...")
    try:
        ticker = yf.Ticker(symbol)

        # Dados diÃ¡rios entre dez anos atrÃ¡s e 2025-05-09
        daily_decade = ticker.history(interval="1d", start=dez_anos_atras, end=data_limite + timedelta(days=1))

        # Estrutura final
        stock_data = {
            "symbol": symbol,
            "daily_last_10y": df_to_records(daily_decade)
        }

        save_stock_data(symbol, stock_data)
        print(f"âœ… Sucesso: {symbol}")

    except Exception as e:
        print(f"âŒ Erro em {symbol}: {e}")
        falhas.append(symbol)

# Resumo
print("\nğŸ“Š Coleta finalizada!")
print(f"âœ”ï¸ Sucesso: {len(most_popular_symbols) - len(falhas)} aÃ§Ãµes")
print(f"âŒ Falha: {len(falhas)} aÃ§Ãµes")
if falhas:
    print("âš ï¸ AÃ§Ãµes com falha:", falhas)
```
